```python
# define path to the folder and csv file
img_folder_path = "./data/wrist_val"
csv_file_path = "./data/val.csv"

# define transformations to be applied to images
transform = transforms.Compose(
    [
        transforms.Resize((299, 299)),
        transforms.ToTensor(),
        transforms.Normalize((0.38, 0.38, 0.38), (0.28, 0.28, 0.28)),
    ]
)

# create instance of custom dataset
val_dataset = CustomDataset(img_folder_path, csv_file_path, transform=transform)

# create dataloader for the dataset
batch_size = 64

val_loader = DataLoader(
    val_dataset, batch_size=batch_size, shuffle=False, num_workers=32
)
```

```python
# show a batch of images
for batch_idx, (data, label) in enumerate(val_loader):
    # data shape: (batch_size, channels, height, width)
    # label shape: (batch_size, )

    # show the first image in the batch
#     print(data)
#     # print(data[1].dtype)
    img = data[0]
    img = transforms.ToPILImage()(img)
    img.show()

    # break out of loop after showing first batch of images
    break
```

```python
# Load the saved model weights
trained_model = Inception()
trained_model.load_state_dict(torch.load("wrist_trainedmodel.pt"))
```

```python
from pytorch_lightning import Trainer
# Create a trainer with the required settings
trainer = Trainer(accelerator="gpu", devices=1)

# Make predictions on the new data
predictions = trainer.predict(trained_model, dataloaders=val_loader)
```

```python
predictions = torch.cat(predictions).numpy()
# get the labels from the validation dataset
val_labels = []
for _, labels in val_dataset:
    val_labels.append(labels)
val_array = np.array(val_labels, dtype=np.float32)
```

```python
plt.scatter(val_array, predictions, s=4, label='validation set predictions')
# add a line to the plot
plt.plot([0, 228], [0, 228], color='black')
plt.plot([0, 228], [-6, 222], color='red')
plt.plot([0, 228], [6, 234], color='red')
plt.xlabel('bone age')
plt.ylabel('wrist GP predictions')
plt.legend()
plt.show()
```

```python
measurements1 = val_array
measurements2 = predictions.squeeze()


# Compute the differences and means
differences = measurements1 - measurements2
means = (measurements1 + measurements2) / 2

# Calculate the mean difference and the limits of agreement
mean_difference = np.mean(differences)
sd_difference = np.std(differences)
lower_limit = mean_difference - 1.96 * sd_difference
upper_limit = mean_difference + 1.96 * sd_difference

# Plot the data and the limits of agreement
plt.scatter(means, differences)
plt.axhline(y=mean_difference, color='gray', linestyle='--')
plt.axhline(y=lower_limit, color='red', linestyle='--')
plt.axhline(y=upper_limit, color='red', linestyle='--')
plt.axhline(y=6, color='green', linestyle='--')
plt.axhline(y=-6, color='green', linestyle='--')

# Add labels and title
plt.xlabel('Mean of AI and reference bone ages')
plt.ylabel('Difference between AI and reference bone ages')
plt.title('Bland-Altman Plot')

# Show the plot
plt.show()
```
